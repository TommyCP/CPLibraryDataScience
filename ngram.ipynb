# Ngram calculator
This notebook reads in a text file and calculates the number of tokens, bigrams, and trigrams. It also sorts them according to frequency.

```python
with open("ddos.txt") as file:
     data = file.read().replace('\n','')
tokens = data.split()
```

```python
bigram = list(zip(tokens,tokens[1::]))
```

```python
trigram = list(zip(tokens,tokens[1::],tokens[2::]))
```

```python
from collections import Counter
```

```python
Counter(tokens).most_common()
```

```python
Counter(bigram).most_common()
```

```python
Counter(trigram).most_common()
```
